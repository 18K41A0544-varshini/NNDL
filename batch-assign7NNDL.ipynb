```python
import pandas as pd
data={'x':[0.2,0.4,0.6,0.8],
      'y':[3.4,3.8,4.2,4.6]}
df=pd.DataFrame(data)

print(df['x'])
```

```python
it=0
max_it=100
n=0.1
m=1
c=-1
s=0
ns=4
deltam=1
deltac=1
derm=1
sum1=0
sum2=0
n1=0

while it<max_it:
    i=0
    while n1<ns:
        for i in range(len(df)):
            m1=m
            c1=c
            one=(df.iloc[i]['y'])
            two=(df.iloc[i]['x'])
            sum1=sum1+((one-(m1*two)-c1)*two)
            sum2=sum2+(one-(m1*two)-c1)
          # print(sum1)
           # print(sum2)
            n1=n1+1
        i=i+1
            
    derm=-(1/ns)*sum1
    deltam=-(n)*derm
    m=m+deltam
    derc=-(1/ns)*sum2
    deltac=-(n)*derc
    c=c+deltac
    
    print("value of m is",m)
    print("value of c is",c)
    it=it+1
    print("iteration",it)
    

print("m value is",m)
print("c value is",c)
            
```
