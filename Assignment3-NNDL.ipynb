```python
import pandas as pd
data={'x':[0.2,0.4,0.6,0.8],
      'y':[3.4,3.8,4.2,4.6]}
df=pd.DataFrame(data)

print(df['x'])
```

```python
it=0
max_it=100
n=0.1
m=1
c=-1
s=0
ns=4
deltam=1
deltac=1
derm=1


while it<max_it:
    s=0
    i=0
    while s<ns:
        for i in range(len(df)):
            m1=m
            c1=c
            one=(df.iloc[i]['y'])
            two=(df.iloc[i]['x'])
            derm=-((one-(m1*two)-c1)*two)
            deltam=-n*derm
            m=m+deltam
            derc=-(one-(m1*two)-c1)
            deltac=-n*derc
            c=c+deltac
            i=i+1
            print("value of m is",m)
            print("value of c is",c)
            s=s+1
        it=it+1
        print("iteration",it)
    

print("m value is",m)
print("c value is",c)
            
```
