```python
import pandas as pd
data={'x':[0.2,0.4,0.6,0.8],
      'y':[3.4,3.8,4.2,4.6]}
df=pd.DataFrame(data)

print(df['x'])
```

```python
it=0
max_it=100
n=0.1
m=1
c=-1
s=0
ns=4
deltam=1
deltac=1
#derm=1
Gm=0
Gc=0
e=10**-8


while it<max_it:
    s=0
    i=0
    while s<ns:
        for i in range(len(df)):
            one=(df.iloc[i]['y'])
            two=(df.iloc[i]['x'])
            gm=-(one-(m*(two))-(c))*two
            print("gm",gm)
            #vm=((r*vm)-(n*gm))
           # print("vm",vm)
            Gm=Gm+(gm*gm)
            print("Gm",Gm)
            deltam=((-n)/((Gm+e)**(1/2)))*gm
            m=m+deltam
            gc=-(one-(m*(two))-(c))
            print("gc",gc)
           # vc=(r*vc)-(n*gc)
           # print("vc",vc)
            Gc=Gc+(gc*gc)
            print("Gc",Gc)
            deltac=((-n)/((Gc+e)**(1/2)))*gc
            c=c+deltac
            i=i+1
            print("value of m is",m)
            print("value of c is",c)
            s=s+1
        it=it+1
        print("iteration",it)
    

print("m value is",m)
print("c value is",c)
            
```
